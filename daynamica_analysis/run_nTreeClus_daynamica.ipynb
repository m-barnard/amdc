## Load packages and source file

```python
import numpy as np
import pandas as pd
import os
import itertools
from itertools import repeat
import sys
sys.path.insert(1,  f'{os.getcwd()}/src/run_functions')
import ntrees_edit
from ntrees_edit import nTreeClus
```

## Run day sequence analysis

```python
df = pd.read_csv(f'clean_data/clean_day_seqs.csv')
sequences = df.loc[:,'seqs'].tolist()
model     = nTreeClus(sequences, n = None, ntree=10, method = "All", max_clust = 9, verbose = 0)
model.nTreeClus()
out = model.output()
clust_res_RF = out['labels_RF'].tolist()
clust_res_RF_p = out['labels_RF_p'].tolist()
df['ntrees_RF'] = clust_res_RF
df['ntrees_RF_p'] = clust_res_RF_p
df.head()
df.to_csv(f'results/amdc_ntrees.csv')
```

## Run week sequence analysis

```python
df2 = pd.read_csv(f'clean_data/clean_week_seqs.csv')
sequences2 = df2.loc[:,'seqs'].tolist()
model2     = nTreeClus(sequences2, n = None, ntree=10, method = "All", max_clust = 9, verbose = 0)
model2.nTreeClus()
out2 = model2.output()
clust_res_DT = out2['labels_DT'].tolist()
df2['ntrees_DT'] = clust_res_DT
df2.head()
df2.to_csv(f'results/amdc_week_ntrees.csv')
```
